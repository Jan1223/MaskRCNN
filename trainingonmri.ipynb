```python
import os
import numpy as np
import skimage.io
import tensorflow as tf
from mrcnn.config import Config
from mrcnn import model as modellib, utils
import matplotlib.pyplot as plt
```

1. Configuration

```python
class BrainMRIMaskRCNNConfig(Config):
    NAME = "brain_mri"
    GPU_COUNT = 1
    IMAGES_PER_GPU = 1
    NUM_CLASSES = 1 + 1 # background + tumor
    IMAGE_MIN_DIM = 256
    IMAGE_MAX_DIM = 256
    STEPS_PER_EPOCH = 42
    DETECTION_MIN_CONFIDENCE = 0.9

config = BrainMRIMaskRCNNConfig()
```

2. Dataset Loader

```python
class BrainMRIDataset(utils.Dataset):
    def load_mri(self, dataset_dir, is_validation=False):
        self.add_class("brain", 1, "tumor")

        for filename in os.listdir(dataset_dir):
            if filename.endswith("_flair.png"):
                pid = filename.split("_")[2]  # Extract patient ID like 001
                prefix = "BraTS20_Validation_" if is_validation else "BraTS20_Training_"

                t2 = os.path.join(dataset_dir, f"{prefix}{pid}_t2.png")
                flair = os.path.join(dataset_dir, f"{prefix}{pid}_flair.png")
                t1ce = os.path.join(dataset_dir, f"{prefix}{pid}_t1ce.png")
                mask = os.path.join(dataset_dir, f"{prefix}{pid}_mask.png") if not is_validation else None

                self.add_image(
                    source="brain",
                    image_id=pid,
                    path=t2,
                    extra={"flair": flair, "t1ce": t1ce, "mask": mask}
                )

    def load_mask(self, image_id):
        info = self.image_info[image_id]
        mask_path = info['extra']['mask']
        mask = skimage.io.imread(mask_path)
        mask = mask.astype(bool)

        # Ensure the mask has a single channel (height, width, 1)
        return np.expand_dims(mask, axis=-1), np.array([1], dtype=np.int32)

    def load_image(self, image_id):
        info = self.image_info[image_id]
        t2 = skimage.io.imread(info['path'])
        flair = skimage.io.imread(info['extra']['flair'])
        t1ce = skimage.io.imread(info['extra']['t1ce'])

        # Ensure images are 2D grayscale before stacking
        print(f"Shape of T2: {t2.shape}")
        print(f"Shape of Flair: {flair.shape}")
        print(f"Shape of T1CE: {t1ce.shape}")

        # Stack the 3 images along the third axis (RGB)
        rgb = np.stack([t2, flair, t1ce], axis=-1).astype(np.uint8)

        print(f"Stacked RGB image shape: {rgb.shape}")
        if rgb.shape[-1] != 3:
            raise ValueError(f"Expected 3 channels, but got {rgb.shape[-1]} channels.")

        return rgb
```

5. Inference and Saving Predicted Masks

```python
model_infer = modellib.MaskRCNN(mode="inference", config=config, model_dir=MODEL_DIR)
model_path = model.find_last()
model_infer.load_weights(model_path, by_name=True)

# Create output dir
OUTPUT_DIR = os.path.join(ROOT_DIR, "predicted_masks")
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Loop through validation images
for image_id in val_dataset.image_ids:
    image = val_dataset.load_image(image_id)
    pid = val_dataset.image_info[image_id]['id']
    result = model_infer.detect([image], verbose=0)[0]

    if result['masks'].shape[-1] > 0:
        combined_mask = np.any(result['masks'], axis=-1).astype(np.uint8) * 255
    else:
        combined_mask = np.zeros(image.shape[:2], dtype=np.uint8)

    save_path = os.path.join(OUTPUT_DIR, f"{pid}_predicted_mask.png")
    skimage.io.imsave(save_path, combined_mask)
    print(f"[INFO] Saved mask: {save_path}")
```
